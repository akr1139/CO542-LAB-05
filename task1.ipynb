```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
#from sklearn.metrics import plot_confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
#train data
data = [[0,0] , [0,1] , [1,0] , [1,1]]
labels = [0,1,1,0]
```

```python
#visualizing the data
sct = plt.scatter([data[0] for data in data], [data[1] for data in data], c=labels)
plt.legend(handles=sct.legend_elements()[0],labels=["0", "1"])
plt.xlabel('input 0')
plt.ylabel('input 1')
plt.title("XOR gate")
plt.show()
```

```python
# define the MLP
netXOR = MLPClassifier(hidden_layer_sizes=(5),activation="relu",random_state=1,max_iter=100)

#train the MLP
clf = netXOR.fit(data, labels)
```

```python
#Accuracy socre
print(f"Accuracy score : {clf.score(data, labels)}")
#prediction
pred = clf.predict([[0,0]])
print(f"predicting [0,0] --> {pred}")

#prediction
pred = clf.predict([[0,1]])
print(f"predicting [0,1] --> {pred}")

#prediction
pred = clf.predict([[1,0]])
print(f"predicting [1,0] --> {pred}")

#prediction
pred = clf.predict([[1,1]])
print(f"predicting [1,1] --> {pred}")
```

```python
# Importing necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Creating the neural network model
netXOR = Sequential()

# Adding layers to the model
# Input layer with 2 neurons
netXOR.add(Dense(units=5, input_dim=2, activation='relu'))  # Hidden layer with 5 neurons and ReLU activation
netXOR.add(Dense(units=1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation

# Compiling the model
netXOR.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training the model with XOR data
# Input data
X_train = [[0, 0], [0, 1], [1, 0], [1, 1]]
# Corresponding target data
y_train = [0, 1, 1, 0]

# Training the model
netXOR.fit(X_train, y_train, epochs=1000, verbose=1)

# Testing the model
test_data = [[0, 0], [0, 1], [1, 0], [1, 1]]
predictions = netXOR.predict_classes(test_data)

# Printing predictions
print("Predictions:")
for i in range(len(test_data)):
    print(f"Input: {test_data[i]}, Predicted Output: {predictions[i][0]}")
```
