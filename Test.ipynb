```python
import numpy as np
import pandas as pd
# Method 1 : Requires Model
from sklearn.metrics import *
# Method 2: Requires Actual and Predicted Features
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Build Logistic Regression Model
from sklearn.linear_model import LogisticRegression
# train test partition
from sklearn.model_selection import train_test_split

# Get Data Import Path
import os
for dirname, _, filenames in os.walk ('/kaggle/input'):
  for filename in filenames:
    print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/content/heart_cleveland_upload.csv')
data.head()
```

```python
y=data['condition']
X=data.drop('condition',axis=1)
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

Logistic regression

```python
lr=LogisticRegression(max_iter=1000,random_state=42)
```

fit model and Train data

```python
lr.fit(X_train,y_train)
```

Make prediction on Test data

```python
y_test_predicted=lr.predict(X_test)
y_test_predicted
```

Use model and test data

```python
import matplotlib.pyplot as plt
from scikitplot.metrics import plot_confusion_matrix


# Plot the confusion matrix
plot_confusion_matrix(y_test, lr.predict(X_test), title='Confusion Matrix')
plt.show()
```
